```python
#Task 1 — File Read & Write Basics 
from pathlib import Path

base = Path.cwd() # we set the Current Working Directory as base
# part A
# 1. to write in the file, we create and open a file called python_notes. we use 'w' as a command to write
text_path = base / "python_notes.txt"
with open(text_path, "w", encoding="utf-8") as f:
    f.write("Topic 1: Variables store data. Python is dynamically typed.\n")
    f.write("Topic 2: Lists are ordered and mutable.\n")
    f.write("Topic 3: Dictionaries store key-value pairs.\n")
    f.write("Topic 4: Loops automate repetitive tasks.\n")
    f.write("Topic 5: Exception handling prevents crashes.\n")
print("File written successfully")


# 2. to add more lines to the file we open that file and use append 'a' command
with open(text_path, "a", encoding="utf-8") as f:
    f.write("Python was named after the show 'Monty Python's Flying Circus'\n")
    f.write("Python was developed by Guido Van Rossum.\n")
print("Lines Appended")

# part B
# 1. to read from the file, we open the file and use the read 'r' command. using enumerate helps to number each line in the file.
# in this particular bit of code, we are printing the lines of the text one by one
count_lines = 0
print("--- File content (line by line) ---")
with open(text_path, "r", encoding="utf-8") as f:
    for number, line in enumerate(f, start=1):
        print(f"{number}:", line)
        count_lines += 1

print(f"\nTotal number of lines in file: {count_lines}")


# to find words the user is looking for, we open the file and run a loop to search the word.
search_text = input("Enter the word you want to search: ")
found = False
with open(text_path, "r", encoding="utf-8") as f:
    for line in f:
        if search_text.lower() in line.lower():
            found = True
            print(line)
if not found:
    print("The word you are searching for is not found.")
```

```python
#Task 2 — API Integration
import requests

BASE = "https://dummyjson.com"

def just_for_fun(title, r):
    print("\n==============================")
    print(title)
    print("==============================")
    print("Status:", r.status_code)
    print("URL:", r.url)
    print("Response JSON:")
    print(r.json())


# to fetch 20 products using get request
resp_20 = requests.get(f"{BASE}/products?limit=20")
# we get the data and store it in json format
data = resp_20.json()
products = data["products"]

#to print the table header
print(f"{'ID':<4} | {'Title':<30} | {'Category':<15} | {'Price':<8} | {'Rating'}")
print("-" * 75)
    
for p in products: # we loop through the products json and add it to the table
    print(f"{p['id']:<4} | {p['title'][:30]:<30} | {p['category']:<15} | ${p['price']:<7.2f} | {p['rating']}")

fands = [] # fands here is just a variable name i have used, means filtered and sorted
for item in products: # to filter out products with rating grater that 4.5/5
    if item["rating"] >= 4.5:
        fands.append(item)

def sort_by_price(item):
    return item["price"]

fands.sort(key=sort_by_price, reverse=True) # to sort the products that we obtained, in ascending order by price
print()
print(fands) # we append to the empty list

print("\nProducts with rating above 4.5 cand sorted by price low to high")
for item in fands:
    print(f"{item['title']} | ${item['price']} | Rating: {item['rating']}")
print()
# to look for a particular category 'laptops
look_for = "laptops"
resp_search = requests.get(f"{BASE}/products/category/{look_for}")
nbs = resp_search.json() # nbs is another variable, i named it as nbs for 'notebooks'
ltp_list = nbs["products"]
for ltp in ltp_list:
    name = ltp["title"]
    price = ltp["price"]
    print(f"Laptop Name: {name:<30} | Price: ${price}")
# to send a new product to the server using post request
new_product = {
  "title": "My Custom Product",
  "price": 999,
  "category": "electronics",
  "description": "A product I created via API"
}

resp_post = requests.post(f"{BASE}/products/add", json=new_product)
just_for_fun("CREATE NEW PRODUCT (POST)", resp_post)
```

```python
# Task 3 — Exception Handling
# Part A — Guarded Calculator
def safe_divide(a, b): #here we define a function to divide two numbers with throwing exceptions
    try:
        ans = a / b
        return ans
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"
    except TypeError:
        return "Error: Invalid input types"
    
print(safe_divide(10, 2))
print(safe_divide(10, 0))
print(safe_divide("ten", 2))
```

```python
# Part B — Guarded File Reader
def read_file_safe(filename): # we define a function to open a particular file
    try:
        with open(filename, 'r') as f:
            return f.read()
    except FileNotFoundError:
        return f"Error: File '{filename}' not found."
    finally:
        print("File operation attempt complete.")

print(read_file_safe("python_notes.txt")) # this will only work if the file we created earlier 'python_notes.txt' exists in the same directory as this file, and not in another folder
print(read_file_safe("games.txt"))
```

```python
#Part C — Robust API Calls
# updated the code with try-except blocks
import requests
def get_20(): # we define function to get 20 products. but we also use try-except block so that execution of the code should not stop.
    try:
        return requests.get(f"{BASE}/products?limit=20", timeout=5)
    except requests.exceptions.ConnectionError:
        return "Connection failed. Please check your internet."
    except requests.exceptions.Timeout:
        return "Request timed out. Try again later."
    except Exception as e:
        return f"Unexpected Error: {e}"

def get_look_for(look_for): # similarly for looking for a particular product or category the user wants to look for
    try:
        return requests.get(f"{BASE}/products/category/{look_for}", timeout=5)
    except requests.exceptions.ConnectionError:
        return "Connection failed. Please check your internet."
    except requests.exceptions.Timeout:
        return "Request timed out. Try again later."
    except Exception as e:
        return f"Unexpected Error: {e}"

def post_new_prod(product_data): # similarly to post a new product and it's details to the server. a mistake in posting or fetching data should not stop entire processes
    try:
        return requests.post(f"{BASE}/products/add", json=product_data, timeout=5)
    except requests.exceptions.ConnectionError:
        return "Connection failed. Please check your internet."
    except requests.exceptions.Timeout:
        return "Request timed out. Try again later."
    except Exception as e:
        return f"Unexpected Error: {e}"


BASE = "https://dummyjson.com"

resp_20 = get_20() # we call the get_20 function here to get 20 products
data = resp_20.json()
products = data["products"]

print(f"{'ID':<4} | {'Title':<30} | {'Category':<15} | {'Price':<8} | {'Rating'}")
print("-" * 75)
    
for p in products:
    print(f"{p['id']:<4} | {p['title'][:30]:<30} | {p['category']:<15} | ${p['price']:<7.2f} | {p['rating']}")

fands = []
for item in products:
    if item["rating"] >= 4.5:
        fands.append(item)

def sort_by_price(item):
    return item["price"]

fands.sort(key=sort_by_price, reverse=True)
print()
print(fands) 

print("\nProducts with rating above 4.5 cand sorted by price low to high")
for item in fands:
    print(f"{item['title']:<30} | ${item['price']:<7.2f} | Rating: {item['rating']}")

resp_search = get_look_for("laptops") # we call the get_look_for function here to get the information about specific item/category the user wants to look for
print(resp_search)
nbs = resp_search.json()
ltp_list = nbs["products"]
for ltp in ltp_list:
    name = ltp["title"]
    price = ltp["price"]
    print(f"Laptop Name: {name:<30} | Price: ${price}")

new_product = {
  "title": "My Custom Product",
  "price": 999,
  "category": "electronics",
  "description": "A product I created via API"
}

resp_post = post_new_prod(new_product) # here we call the post_new_prod function to add a new product and it's details to the server
print(resp_post.json())
```

```python
#Part D — Input Validation Loop:
while True: # we start an infinite loop
    prod_id = input("Enter a product ID to look up (1 to 100), or 'quit' to exit: ") # we ask for input from the user
    if prod_id.lower() == 'quit': # if the user types in 'quit' break the loop
        break
    if not prod_id.isdigit(): # here we check whether the value entered by the user is valid or not.
        print("The value you enter should be integer, not text or decimals")
        continue
    prod_num = int(prod_id) # the value entered by the user is still in string. so we convert it to integer. but the number/value entered by the user must also be between 1 and 100
    if not 1 <= prod_num <= 100:
        print("The values should be between 1 and 100 only.")
        continue
    try: # try to get the product the user asked for 
        prod_url = f"https://dummyjson.com/products/{prod_num}"
        resp_prod_id = requests.get(prod_url, timeout=5)
        if resp_prod_id.status_code == 404:
            print("Product not found")
        elif resp_prod_id.status_code == 200:
            prod = resp_prod_id.json()
            print(f"{prod['title']} | ${prod['price']}")
    except Exception as e:
        print(f"Unexpected Error: {e}")
```

```python
#Task 4 — Logging to File
import requests
from datetime import datetime

def log_error(function_name, error_type, message): # we define a function to automatically open the error logs file and append to it any errors
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S") # the timestamp in the format we want
    log_entry = f"[{timestamp}] ERROR in {function_name}: {error_type} — {message}\n" # here we formatting the entries of the error log file
    with open("error_log.txt", "a") as f: # it will open the file and append automatically
        f.write(log_entry)

# catching a connection error
try:
    wrong_url = "https://this-host-does-not-exist-xyz.com/api" # we purposely give a URL that does not exist. so if a call is made to this link, there is nothing to fetch because the server does not even exist
    resp = requests.get(wrong_url, timeout=2) # this would trigger an error and python will stop executing anything further. to stop this, we use try-except block
except requests.exceptions.ConnectionError as e:
    log_error("fetch_products", "ConnectionError", "Connection could not be made.")
except Exception as e:
    log_error("fetch_products", "UnexpectedError", e)

# to trigger a error 404
invalid_id = 999 # in this case, a connection is made to the server, the server can only send back an invalid code error. hence we do not use try-except block.
url_404 = f"https://dummyjson.com/products/{invalid_id}"
resp_404 = requests.get(url_404)

if resp_404.status_code != 200:
    log_msg = f"{resp_404.status_code} Not Found for product ID {invalid_id}"
    log_error("lookup_product", "HTTPError", log_msg)

# to display the error log, we read the file. if the file is not created yet or does not exist, it throws an error
try:
    with open("error_log.txt", "r") as f:
        print(f.read())
except FileNotFoundError:
    print("Log file not created yet.")
```
